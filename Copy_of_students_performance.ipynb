```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('Students_Performance_data_set.xlsx')
df = df.rename(columns={
    'Gender': 'gender',
    'How many hour do you spent daily in social media?': 'social_hours',
    'How many hour do you study daily?': 'study_hours',
    'What is your current CGPA?': 'CGPA'
})

# ===== 1. MALE vs FEMALE: Social Media Impact =====
print("Correlation by Gender:")
for g in df['gender'].unique():
    subset = df[df['gender'] == g]
    corr = subset['social_hours'].corr(subset['CGPA'])
    print(f"{g}: {corr:.3f}")

# Plot: Avg CGPA vs Social Hours, split by Gender
plt.figure(figsize=(8,5))
df.groupby(['social_hours','gender'])['CGPA'].mean().unstack().plot(kind='bar')
plt.title("Avg CGPA vs Social Media Hours by Gender")
plt.xlabel("Social Media Hours"); plt.ylabel("Avg CGPA")
plt.legend(title="Gender")
plt.xticks(rotation=0)
plt.show()

# ===== 2. SOCIAL vs STUDY: Which affects CGPA more? =====
corr_social = df['social_hours'].corr(df['CGPA'])
corr_study = df['study_hours'].corr(df['CGPA'])

print(f"\nCorrelation with CGPA:")
print(f"Social Media Hours: {corr_social:.3f}")
print(f"Study Hours: {corr_study:.3f}")

# Plot both side by side
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(df['social_hours'], df['CGPA'], alpha=0.5, color='red')
z = np.polyfit(df['social_hours'], df['CGPA'], 1)
plt.plot(df['social_hours'].sort_values(), np.poly1d(z)(df['social_hours'].sort_values()), "r--")
plt.title("Social Hours vs CGPA"); plt.xlabel("Social Hours"); plt.ylabel("CGPA")

plt.subplot(1,2,2)
plt.scatter(df['study_hours'], df['CGPA'], alpha=0.5, color='green')
z = np.polyfit(df['study_hours'], df['CGPA'], 1)
plt.plot(df['study_hours'].sort_values(), np.poly1d(z)(df['study_hours'].sort_values()), "g--")
plt.title("Study Hours vs CGPA"); plt.xlabel("Study Hours"); plt.ylabel("CGPA")

plt.tight_layout()
plt.show()
```

Gender split: If Female correlation is -0.4 and Male is -0.1, social media hits females harder
Social vs Study: study_hours should have positive correlation social_hours usually has negative correlation

Whichever number is bigger in absolute value = bigger impact

FINIDNGS: |corr_study| > |corr_social| → Study time matters more than social media time.
